# Slow Python data structures

## Lists and loops

__List__ and __dictionaries__ are the main Python data structures. 

This is a list:

```python
numbers = [1, 2, 3, 4, 5]
```

Lists are accessed by index:

```python
numbers[2]
```

This is a dictionary:

```python
dictio = {"name": "Jin", "age": 42}
```

Dictionaries are accessed by key:

```python
dictio["name"]
```

Lists and dictionaries are often manipulated with for loops:

```python
def create_numbers(n_values):
    numbers = []
    for x in range(n_values):
        numbers += [1 + x]
    return numbers
```

```python
numbers = create_numbers(1000)
numbers
```

## Optimizing for loops

Python loops are slow:

```python
%%timeit
inverses = []
for x in numbers:
    inverses += [1 / x]
```

List comprehensions reduce the Python interpreter overhead:


```python
%%timeit
inverses_comp = [1 / x for x in numbers]
```

# Fast NumPy arrays

## NumPy overview
NumPy arrays are the "workhorse" of data analysis. They are omnipresent in data science.
NumPy is fast because it is partially implemented in C, which provides for more optimized data structures.

A NumPy array can be created in various ways, including from a Python list:

```python
import numpy as np

numbers_np = np.array(numbers)
numbers_np
```

NumPy provides a number of __vectorized__ functions that apply to arrays, such as:

```python
# basic arithmetic
numbers_np + 1
```

```python
# Mathematical functions
np.sqrt(numbers_np)
```

```python
# sum, count, min, max, etc
np.sum(numbers_np)
```

# Scipy
* Scientific computation library built upon Numpy
* Numpy provides basic functions while Scipy provides mathematical functions


**Differentiation**

```python
def f(x):
    return x**2 * np.sin(2 * x) * np.exp(-x)


x = np.linspace(0, 1, 100)

plt.plot(x, f(x));
```

```python
from scipy.misc import derivative

plt.plot(x, f(x))
plt.plot(x, derivative(f, x, dx=1e-5));
# plt.plot(x, derivative(f, x, dx=1e-5, n=2));
```

**Linear Algebra**

```python
from scipy.linalg import solve_triangular

a = np.array([[3, 0, 0, 0], [2, 1, 0, 0], [1, 0, 1, 0], [1, 1, 1, 1]])
b = np.array([3, 4, 3, 4])

solve_triangular(a, b, lower=True)
```

**Statistics**

```python
from scipy.stats import norm

# normal distribution
mu = 0
sigma = 1
dist = norm(mu, sigma)
```

```python
# Get the probability density function (pdf) of the Normal Distribution
x = np.linspace(-10, 10, 100)
pdf = dist.pdf(x)

plt.plot(x, pdf);
```
